# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Berikut adalah tahapan-tahapan yang bisa dilakukan, tetapi **tidak terbatas** pada:
1. Menghapus atau Menangani Data Kosong (Missing Values)
2. Menghapus Data Duplikat
3. Normalisasi atau Standarisasi Fitur
4. Deteksi dan Penanganan Outlier
5. Encoding Data Kategorikal
6. Binning (Pengelompokan Data)

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
# Cek & Isi Missing Values
# Jika ada yang kosong, isi dengan median (untuk angka)
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

print("Pengecekan missing values selesai.")
print(df.isnull().sum())
```

```python
# Cek & Hapus Duplikat
duplicate_count = df.duplicated().sum()
if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print(f"{duplicate_count} data duplikat dihapus.")
else:
  print("Tidak ada data duplikat.")
```

```python
# Handling outlier
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Kolom yang sering memiliki nilai ekstrem
cols_to_check = ['Apps', 'Accept', 'Enroll', 'Expend']
print(f"Data sebelum filter outlier: {len(df)}")
df = remove_outliers(df, cols_to_check)
print(f"Data setelah filter outlier: {len(df)}")
```

```python
# Fitur Engineering
# Menghitung Acceptance Rate (Berapa persen yang diterima dari total pendaftar)
df['Accept_Rate'] = (df['Accept'] / df['Apps']) * 100

print("Fitur 'Accept_Rate' berhasil dibuat.")
display(df[['Apps', 'Accept', 'Accept_Rate']].head())
```

```python
# Binning Grad.Rate menjadi 3 kategori
labels = ['Low_Grad_Rate', 'Medium_Grad_Rate', 'High_Grad_Rate']
df['Grad_Category'] = pd.qcut(df['Grad.Rate'], q=3, labels=labels)

print("Binning Grad.Rate selesai.")
display(df[['Grad.Rate', 'Grad_Category']].head())
```

```python
# Standarisasi (StandardScaler)
# Untuk data dengan rentang luas: Apps, Accept, Enroll, Outstate, Expend
std_scaler = StandardScaler()
cols_to_std = ['Apps', 'Accept', 'Enroll', 'Outstate', 'Expend', 'Room.Board']
df[cols_to_std] = std_scaler.fit_transform(df[cols_to_std])

# Normalisasi (MinMaxScaler)
# Untuk persentase/rasio: PhD, S.F.Ratio, Grad.Rate, Accept_Rate
mm_scaler = MinMaxScaler()
cols_to_norm = ['PhD', 'S.F.Ratio', 'Grad.Rate', 'Accept_Rate', 'perc.alumni']
df[cols_to_norm] = mm_scaler.fit_transform(df[cols_to_norm])

print("Penskalaan data selesai.")
df.head()
```

```python
# Mengubah kolom 'Private' (Yes/No) dan 'Grad_Category' menjadi angka
categorical_cols = ['Private', 'Grad_Category']

# One-hot encoding
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=int)

print("Encoding selesai. Semua kolom kini sudah numerik.")
display(df.head())
```
